```python
sns.set(font_scale=1)
f, ax = plt.subplots(3, 3, figsize=(10, 10), sharex=False, sharey=False)

#The distribution of Calories and Colesterol
x = df['Calories'].values
y = df['Cholesterol (% Daily Value)'].values
sns.kdeplot(x, y, ax=ax[0, 0])
sns.rugplot(x, color='g', ax=ax[0, 0])
sns.rugplot(x, vertical=True, ax=ax[0, 0])
ax[0, 0].set(xlim=(-200, 1000), ylim=(-10, 40),  title = 'Calories and Cholesterol')

#The distribution of Total Fat and Colesterol  
x = df['Total Fat (% Daily Value)'].values
y = df['Cholesterol (% Daily Value)'].values
sns.kdeplot(x, y, ax=ax[0, 1])
sns.rugplot(x, color='g', ax=ax[0, 1])
sns.rugplot(x, vertical=True, ax=ax[0, 1])
ax[0, 1].set(xlim=(-20, 70), ylim=(-10, 40), title = 'Total Fat and Cholesterol')
```

```python
trace1 = go.Scatter(
    x = df['Total Fat (% Daily Value)'],
    y = df['Calories'],
    name = 'TotalFat/Calories',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'rgba(152, 0, 0, .8)',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    )
)

trace2 = go.Scatter(
    x = df['Protein'],
    y = df['Calories'],
    name = 'Protein/Calories',
    mode = 'markers',
    marker = dict(
        size = 10,
        color = 'rgba(205, 160, 193, .9)',
        line = dict(
            width = 2,
            color = 'rgb(0, 0, 0)'
        )
    )
)


data = [trace1, trace2]
layout = dict(title = 'The Influence of Total Fat & Protein on Calories',
              yaxis = dict(zeroline = False),
              xaxis = dict(zeroline = False)
             )

fig = dict(data=data, layout=layout)
py.iplot(fig, filename='styled-scatter')
```

```python
trace1 = go.Scatter3d(
    x=df['Category'].values,
    y=df['Item'].values,
    z=df['Cholesterol (% Daily Value)'].values,
    text=df['Item'].values,
    mode='markers',
    marker=dict(
            size=12,
            color = df['Cholesterol (% Daily Value)'].values,
            colorscale = 'Portland',
            colorbar = dict(title = 'Total Fat (% Daily Value)'),
            line=dict(color='rgb(255, 255, 255)'),
    )
)
data = [trace1]
layout = go.Layout(
    margin=dict(
        l=0,
        r=0,
        b=0,
        t=0
    )
)
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='simple-3d-scatter')
```
